# ***DATA ANALYTICS***

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df=pd.read_csv('/kaggle/input/nba-active-players-data-images/players.csv')
df
```

```python
# HISTOGRAM 
```

```python
#PIE CHART
country=df['country'].unique()
print(country)
```

```python
# country_player_count=np.zeros(len(country))
country_player_count=Counter(df['country'])
print(country_player_count)
```

```python
# Create a pie chart using Matplotlib with dark mode
plt.figure(figsize=(8, 8))
plt.style.use('dark_background')  # Use dark mode style

# Customize the colors for the pie chart
colors = plt.cm.Dark2.colors
plt.pie(country_player_count.values(), labels=country_player_count.keys(), autopct='%1.1f%%', startangle=140, colors=colors)

plt.title('Pie Chart of Country Player Counts (Dark Mode)')
plt.show()
```

```python
# Set the style and color palette for Seaborn
sns.set(style="darkgrid")
sns.set_palette("dark")

# Convert the counter data to a list of tuples (needed for countplot)
count_data = [(country, count) for country, count in country_player_count.items()]
print(count_data)

# # Create a countplot to simulate a pie chart using Seaborn
# plt.figure(figsize=(8, 8))
# sns.countplot(x=1, data=count_data, palette=sns.color_palette())
# plt.title('Pie Chart of Country Player Counts (Dark Mode)')
# plt.ylabel("")  # Remove y-axis label
# plt.xlabel("")  # Remove x-axis label
# plt.xticks([])  # Remove x-axis tick labels
# plt.show()
```

```python
# BAR GRAPH
df1 = pd.DataFrame(count_data, columns=['Country', 'Count'])

# Create a dark-themed barplot using Seaborn
plt.figure(figsize=(8, 30))
ax = sns.barplot(x='Count', y='Country', data=df1)  # Reverse x and y for horizontal bars

# Customize the appearance
plt.title('Bar Graph of Country Player Counts (Dark Mode)')
plt.xlabel('Count')
plt.ylabel('Country')


plt.show()
```

```python
# Create a figure and axes
plt.figure(figsize=(20, 20))
ax = plt.gca()

# Set dark mode background for both the axes and the figure
ax.set_facecolor('black')
plt.gcf().set_facecolor('black')

# Create the bar graph
countries = [country for country, _ in count_data]
counts = [count for _, count in count_data]

bars = ax.barh(countries, counts, color='darkblue')

# Customize appearance
plt.title('Bar Graph of Country Player Counts (Dark Mode)',color='white')
plt.xlabel('Count',color='white')
plt.ylabel('Country',color='white')
plt.xticks(color='white')  # Customize tick color
plt.yticks(color='white')  # Customize tick color
ax.spines['bottom'].set_color('white')  # Customize axis color
ax.spines['left'].set_color('white')    # Customize axis color

# Add data labels
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width()}',
             va='center', ha='left', color='white', fontsize=10)

plt.show()
```
